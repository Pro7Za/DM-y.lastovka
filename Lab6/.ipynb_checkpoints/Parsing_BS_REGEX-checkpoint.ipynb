#### 1. Аналізуємо вихідний html-код та визначаємось з стратегією вилучення списка підрозділів

#### 2. Будуємо bs-запити для вилучення даних

```python
# вилучаємо стрінку з данними у вигляді utf-8 текста
url = "https://knute.edu.ua/blog/read?n=fmtp&uk"
main_page = requests.get(url)
main_page.encoding = 'utf-8'
main_page_text = main_page.text
```

```python
# будуємо bs-об'єкт
soup = bs(main_page_text)
```

```python
# вилучаємо необхідний контент
fac_list = soup.find_all("ul", class_="nav")[0]      \
                .find_all("li", class_="dropdown-menu")[1] \
                .find_all("a")
```

```python
# вилучаємо елементи які містять 'факультет' або 'кафедра'
names = [fac.string for fac in fac_list if 
        "Факультет " in fac.string or "Кафедра" in fac.string]
```

```python
names[:5]
```

```python
def make_table_dict(names):
    '''функція будує словник де ключ - назва факультету,
    значення - список кафедр'''
    
    table_dict = {}
    for name in names:
        if "Факультет" in name:
            table_dict[name] = []
            current_fac_name = name
        else:
            table_dict[current_fac_name].append(name) 
    return table_dict
```

```python
table_dict = make_table_dict(names)
```

```python
table_dict
```

```python
def make_table(table_dict):
    ''' функція будує датафрейм відповідно до вимог задачі'''
    import pandas as pd
    return pd.DataFrame.from_dict(table_dict, orient="index").transpose()
```

```python
make_table(table_dict)
```

### 3.4. Огляд альтернативних засобів скрапінгу

[__Регулярні вирази__](https://uk.wikipedia.org/wiki/Регулярний_вираз) - це технологія доступу до складових частин строк за допомогою відповідних [правил та шаблонів](https://www.regular-expressions.info/tutorial.html).

```python
import re
print(dir(re))
```

```python
# отримати список всіх тегів з атрибутом 'href'
all_hrefs = re.findall("<a(.*?)<\/a>", main_page_text)
```

```python
all_hrefs[20:30]
```

```python
# вибір елементів в яких присутні 'кафедра' або 'факультет'
import functools as ft
fac_n_kaf_hrefs = ft.reduce(lambda facs, ref_item: facs + [ref_item] 
          if "Факультет " in ref_item or "Кафедра" in ref_item 
          else facs , all_hrefs, [])
```

```python
fac_n_kaf_hrefs[:3]
```

```python
# патерн для вибору тексту на кирилиці
pat = "[А-яіїє]+"
```

```python
proba = fac_n_kaf_hrefs[0]
' '.join(re.findall(pat, proba))
```

```python
# побудова списку підрозділів
names = ft.reduce(lambda names, item: names + [' '.join(re.findall(pat, item))],
          fac_n_kaf_hrefs, [])
names[:10]
```

```python
# побудова словник підрозділів
table_dict = make_table_dict(names)
table_dict
```

```python
# побудова вихідної таблиці
make_table(table_dict)
```
